## Merge

```python
izquierda=pd.DataFrame({"key":["k0","k1","k2","k3"],
"A":["A0","A1","A2","A3"],
"B":["B0","B1","B2","B3"]
 })

derecha=pd.DataFrame({"key":["k0","k1","k2","k3"],
"C":["C0","C1","C2","C3"],
"D":["D0","D1","D2","D3"]
 })
```

```python
#Para obtener un DF con una columna comun
izquierda.merge(derecha,on="key")
```

```python
#Si tuviera nombre distinto la columna con la cual quiero hacer el merge
izquierda=pd.DataFrame({"key":["k0","k1","k2","k3"],
"A":["A0","A1","A2","A3"],
"B":["B0","B1","B2","B3"]
 })

derecha=pd.DataFrame({"key_2":["k0","k1","k2","k3"],
"C":["C0","C1","C2","C3"],
"D":["D0","D1","D2","D3"]
 })

izquierda.merge(derecha,left_on="key",right_on="key_2")
```

## Join - Index match

```python
izq=pd.DataFrame({"A":["A0","A1","A2"],
"B":["B0","B1","B2"]},index=["k0","k1","k2"])

der=pd.DataFrame({"C":["C0","C1","C2"],
"D":["D0","D1","D2"]},index=["k0","k2","k3"])
```

```python
#how= inner
print(izq.join(der,how="inner"))

#how= left
print(izq.join(der,how="left"))

#how=right
print(izq.join(der,how="right"))

#how=outer
print(izq.join(der,how="outer"))
```

## Pivot table

```python

df=pd.read_csv("countries_gdp_hist.csv",";")
```

```python
df.head(5)
```

```python
#Esta función nos permite de cierta forma crear una reorganización de los
#valores de la tabla y asi mismo operar de la forma deseada un determinado valor

df.pivot_table(index="year",columns="region_name",values="total_gdp_million",aggfunc="sum")
```
