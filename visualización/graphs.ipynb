```python
import matplotlib.pyplot as plt 
import numpy as np  
```

```python
data=np.random.randint(1,50,100)
data
```

## Histograma


```python
plt.hist(data,bins=10,histtype="bar") #Bins me define el n√∫mero de barras que quiero, ya que el histograma me muestra frecuencia
plt.show()
```

## Boxplot

```python
data=np.append(data,200)
plt.boxplot(data, vert=False , patch_artist=True, notch=True)
plt.show()
```

## Scatter

```python
N=50
x=np.random.rand(N)
y=np.random.rand(N)
area=(30*np.random.rand(N))**2
colors=np.random.rand(N)
```

```python
plt.scatter(x,y,s=area,c=colors)
plt.show()
```
