```python
import matplotlib.pyplot as  plt    
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=np.sin(x)
```

```python
fig,axes=plt.subplots()
axes.plot(x,y,"b")
```

```python
#Declaro la estructura que quiero en mis subplots
fig,axes=plt.subplots(nrows=1 , ncols=2)
axes[0].plot(x,y,"b") #trabajo en ellos como arreglos
axes[1].plot(y,x,"r")
```

```python
#Por tanto puedo trabajar desde el comienzo como una tupla
fig,(ax1,ax2)=plt.subplots(nrows=1 , ncols=2)
ax1.plot(x,y,"b") #trabajo en ellos como arreglos
ax2.plot(y,x,"r")
```
