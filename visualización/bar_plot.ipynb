```python
import matplotlib.pyplot as  plt
import numpy as np  
```

```python
country=["Col","Ind","Bra","Bol","Ecu"]
population=[1000,3000,5666,2333,4566]
```

```python
plt.bar(country,population,width=0.5,color=["r","green"])# configurando bars
plt.xticks(np.arange(5),("Col","Bra","Bol","Ecu","Ind"),rotation=45)#ordenando los valores del eje x
plt.show()
```

```python
#Tambien se puede hacer horizontal
plt.barh(country,population)
plt.show()
```
