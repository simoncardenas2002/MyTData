```python
import matplotlib.pyplot as  plt
import numpy as np  
```

```python
x= np.linspace(0,5,11)
y= x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,"b")
fig.show()
```

### la ventaja de utilizar orientación a objetos es que puedo cambiar rápidamente las carácterísticas de cada objeto

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2= fig.add_axes([0.1,0.55,0.4,0.3]) #Con estos parámetros logro variar 
# tanto su posición como su tamaño en el espacio

axes.plot(x,y,"b")
axes2.plot(x,y,"r")
axes2.set_facecolor("gray")
fig.show()
```
