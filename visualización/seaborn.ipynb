```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=["A","B","C"],y=[1,3,2])
plt.show()
```

```python
#Customizando la gráfica
sns.set(style="darkgrid",palette="dark",font="verdana",font_scale=1)
sns.barplot(x=["A","B","C"],y=[1,3,2])
plt.show()
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.displot(data=tips, x="total_bill")
plt.show()
```

```python
#Seaborn al ver que el histograma no es una gráfica optima por el tipo de visualización que estoy realizando
# utiliza una visualizacion que si me relacione 2 variables
sns.displot(data=tips, x="total_bill",y="tip",hue="sex")
#hue me permite ademas crear una diferenciación adicional de las variables graficadas
plt.show()
```

```python
#ademas puedo variar la gráfica que me muestra con solo variar kind
sns.displot(data=tips, x="total_bill",y="tip",hue="sex",kind="kde")
plt.show()
```

```python
sns.histplot(data=tips ,x="tip",bins=15)
plt.show()
```

```python
#de forma acumulativa
sns.histplot(data=tips ,x="tip",bins=15,cumulative=True)
plt.show()
```

```python
#asi mismo puedo variar la variable del eje Y, en este caso es frecuencia
sns.histplot(data=tips ,x="tip",bins=15,cumulative=False,hue="sex", stat="frequency")
plt.show()
```

```python
#En este caso es porcentaje
sns.histplot(data=tips,x="tip",bins=15,cumulative=False,hue="sex",stat="percent",multiple="stack")
plt.show()
```

```python
sns.kdeplot(data=tips,x="tip",hue="sex",cumulative=False,shade=True)
plt.show()
```

```python
sns.ecdfplot(data=tips,x="tip",hue="sex",stat="count")
plt.show()
```

```python
sns.displot(data=tips,x="tip",hue="sex",stat="count")
plt.show()
```

##Categócicos

```python
sns.countplot(data=tips,x="day",hue="sex")
plt.show()
```

```python
#stripplot
plt.figure(figsize=(6,4))
sns.stripplot(data=tips,x="day",y="total_bill",hue="sex",dodge=True)
plt.show()
```

```python
#Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(data=tips,x="day",y="total_bill",hue="sex",dodge=True)
plt.show()
```

```python
#Tambien puedo combinar
plt.figure(figsize=(6,4))
sns.boxplot(data=tips,x="day",y="total_bill",hue="sex",dodge=True)

sns.stripplot(data=tips,x="day",y="total_bill",hue="sex",dodge=True,palette="pastel")
plt.show()
```

```python
#Tambien puedo cambiar el parámetro col, para que en la columna se tengan distintas variables
plt.figure(figsize=(6,4))
sns.catplot(data=tips ,x="day",y="total_bill",hue="sex",kind="box",col="time")
#logro trabajar con 3 variables distintas
plt.show()
```

## Relation

```python
#Puedo diferenciar multiples variables
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style="time",size="size")
plt.legend(loc="center",bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
markers={"luch":"D","Dinner":"s"} #Asi logro customizar los markers
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style="time",size="size")
plt.legend(loc="center",bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
sns.lineplot(data=tips,x="total_bill",y="tip")
plt.legend(loc="center")
```
