### #2 muestreo sistemático

```python
def systematic_sampling(data,step):
    indexes=np.arange(0,len(data),step=step) # aqui estoy creando un arreglo de los índices de los cuales quiero sacar la información del dataset
    systematic_sample=data.iloc[indexes] #aca creo un nuevo conjunto de datos basado en esos datos que saqué con los indices anteriores
    return systematic_sample

sample=systematic_sampling(df,3)
sample
```

### #3 muestreo estratificado

```python
#comienzo creando una estratificación de las variables
df["estratificado"]=df["delegacion"]+","+df["tipo"]
df["estratificado"].value_counts()
#para tener estos valores en porcentajes divido por la cantidad de datos
(df["estratificado"].value_counts()/len(df)).sort_values()
```

De lo anterior podemos obtener las primeras conclusiones sobre los datos y las proporciones:
1) Hoteles en Cuantemoc= 0.5
2) Museos en Cuantemoc= 0.2
3) Hoteles en Venustiano= 0.1 


```python
#Para crear la proporcion a partir de los datos se usa la siguiente función que encuentra el número
# de muestras necesarias de cada categoría, hace un muestreo simple dentro de la categoría 
# y luego junta estas muestras
def data_estratificada(df,nombres_columnas_estrat,valores_estrat,prop_estrat,random_state=None):
    df_estrat=pd.DataFrame(columns=df.columns)

    pos=-1
    for i in range(len(valores_estrat)):
        pos +=1
        if pos == len(valores_estrat) -1:
            ratio_len=len(df)-len(df_estrat)
        else:
            ratio_len=int(len(df)*prop_estrat[i])

        df_filtrado=df[df[nombres_columnas_estrat]== valores_estrat[i]]
        df_temp=df_filtrado.sample(replace=True,n=ratio_len,random_state=random_state)

        df_estrat=pd.concat([df_estrat,df_temp])
    return df_estrat

```

```python
#creamos el sampleo
valores_estrat = ['Cuautémoc,Hotel', 'Cuautémoc,Museo', 'Venustiano Carranza,Hotel', 'Cuauhtémoc,Mercado','Venustiano Carranza,Mercado']
prop_estrat = [0.5, 0.2, 0.1, 0.1, 0.1]
df_estrat = data_estratificada(df, 'estratificado', valores_estrat, prop_estrat, random_state=42)
df_estrat
```
