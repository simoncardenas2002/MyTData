```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Leer la base de datos

```python
#Crearemos la base de datos asignando el nomnbre de las columnas que queramos
url="https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names=["sepal-length","sepal-width","petal-length","petal-width","class"]
iris=pd.read_csv(url,names=names)
```

```python
iris.head()
```

```python
plt.hist(iris["sepal-length"],color="orange",bins=30)
```

```python
# a pesar de que anteriormente observamos un comportamiento con cierta 
# tendencia normal, vamos a verificar realmente este comportamiento
sns.distplot(iris["sepal-length"],hist=True,kde=True,bins=30)
```

### Cálculo de varianza

```python
#con pandas puedo obtener la varianza simplemente con la función var 
iris["sepal-length"].var()
```

### Cálculo de la desviación estandar

```python
#asi mismo se tiene la funcion std
iris["sepal-length"].std()
```

Para saber si este dato es significativo o no hallamos el promedio

```python
iris["sepal-length"].mean()
```

concluimos que tenemos una varianza y desviación estandar elevada

```python
## Comenzaremos creando una muestra aleatoria simple
muestra=iris.sample(frac=0.5)
muestra
```

```python
# verificamos las medidas de tendencia anteriores de la muestra
muestra["sepal-length"].var()
```

```python
muestra["sepal-length"].std()
```

```python
muestra["sepal-length"].mean()
```
