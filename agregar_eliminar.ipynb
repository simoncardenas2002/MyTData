## 1) Drop columns

```python
#Asi solo logro hacer drop en la salida
df1.drop("Genre",axis=1).head(2)
```

```python
#Si quiero que se borre del dataframe aplico implace
df1.drop("Genre",axis=1,inplace=True)
#o puedo reasociar el dataframe
df1=df1.drop("Genre",axis=1).head(2)
```

## 2) Drop rows

```python
#Aplica lo mismo que con columnas, simplemente se debe cambiar
# el eje , ahora es 0

df1.drop(0,axis=0).head(2) #aqui borro la fila de indice cero
```

```python
#O si quiero borrar una cantidad de filas determinadas

df1.drop(range(0,10),axis=0).head(2)
```
